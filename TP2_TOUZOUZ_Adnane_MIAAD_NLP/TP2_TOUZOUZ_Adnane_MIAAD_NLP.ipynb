# Exercice 1(Bag of Words)

```python
Review1 =  "Game of Thrones is an amazing tv series!"
Review2 = "Game of Thrones is the best tv series!"
Review3 = "Game of Thrones is so great"
reviews = [Review1, Review2, Review3]
```

```python
reviews
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectoriser = CountVectorizer()
vectoriser.fit(reviews)
```

```python
vectoriser.vocabulary_
```

```python
vectoriser.get_feature_names()
```

```python
reviews_bag_of_words = vectoriser.transform(reviews)
```

```python
reviews_bag_of_words.toarray()
```

# Exercice 3

```python
data = pd.read_csv('SPAM_Data.csv')
```

```python
data
```

```python
# Supprimer les ponctuations
import string
ponc = string.punctuation
liste = []
data["removed_punc"] = ""
for i in range(len(data)):
    text = data['Message'][i]
    liste = [word for word in text if word not in ponc]
    data["removed_punc"][i] = "".join(liste)
data.head(10)
```

```python
import nltk
data['tokens'] = ""
for i in range(len(data)):
    data['tokens'][i] = nltk.word_tokenize(data['removed_punc'][i].lower())
data.head(10)
```

```python
data['larger_tokens'] = ""
for i in range(len(data)):
    liste = [w for w in data['tokens'][i] if len(w) > 3] 
    data['larger_tokens'][i] = liste.copy()
data.head(10)
```

```python
from nltk.corpus import stopwords
stopwords_en = set(stopwords.words('english'))
data['clean_tokens'] = ""
for i in range(len(data)):
    liste = [word for word in data['larger_tokens'][i] if word not in stopwords_en]
    data['clean_tokens'][i] = liste.copy()
data.head(10) 
```

```python
# Stemming (retourner les racines des tokens)
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
data['stem_words'] = ""
for i in range(len(data)):
    liste = [stemmer.stem(word) for word in data['clean_tokens'][i]]
    data['stem_words'][i] = liste
data.head(10)
```

```python
# Lemmatisation (retourner la forme canonique des tokens)
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
data['lemma_words'] = ""
for i in range(len(data)):
    liste = [wnl.lemmatize(word) for word in data['clean_tokens'][i]]
    data['lemma_words'][i] = liste.copy()
data.head(10)
```

```python
data['Clean_text'] = ""
for i in range(len(data)):
    data['Clean_text'][i] = " ".join(data['lemma_words'][i])
data
```

```python
bag_words = []
for i in range(len(data)):
    bag_words.append(data["Clean_text"][i])
bag_words
```

```python
vectoriser.fit(bag_words)
```

```python
vectoriser.vocabulary_
```

```python
vectoriser.get_feature_names()
```

```python
bag_words_final = vectoriser.transform(bag_words)
bag_words_final.toarray()
```
