# EXERCICE 1

```python
import pandas as pd
```

```python
data = pd.read_csv("tweets.csv")
```

```python
data
```

```python
# Supprimer les poncuations
import string
ponc = string.punctuation
for i in range(len(data)):
    data["text"][i] = "".join([w for w in data["text"][i] if w not in ponc])
data
```

```python
# tokinisation en minuscul
import nltk
for i in range(len(data)):
    data["text"][i] = data["text"][i].lower()
    data["text"][i] = nltk.word_tokenize(data["text"][i])
data
```

```python
# Suppresion des stopwords
from nltk.corpus import stopwords
stopwords_en = set(stopwords.words('english'))
for i in range(len(data)):
    data["text"][i] = [w for w in data["text"][i] if w not in stopwords_en]
data
```

```python
# Supprimer les mors inf à 3
for i in range(len(data)):
    data["text"][i] = [w for w in data["text"][i] if len(w)>3]
data
```

```python
# Stemming
from nltk.stem import PorterStemmer
stem = PorterStemmer()
for i in range(len(data)):
    data["text"][i] = [stem.stem(w) for w in data["text"][i]]
data
```

```python
# Clean_text
for i in range(len(data)):
    data["text"][i] = " ".join(data["text"][i])
data
```

```python
# 2-gramme
import re
from nltk.util import ngrams
data["2-gramme"] = ""
for i in range(len(data)):
    data["2-gramme"][i] = [t for t in data["text"][i].split(" ")]
    data["2-gramme"][i] = list(ngrams(data["2-gramme"][i], 2))
data
```

```python
data["3-gramme"] = ""
for i in range(len(data)):
    data["3-gramme"][i] = [t for t in data["text"][i].split(" ")]
    data["3-gramme"][i] = list(ngrams(data["3-gramme"][i], 3))
data
```

```python
data["4-gramme"] = ""
for i in range(len(data)):
    data["4-gramme"][i] = [t for t in data["text"][i].split(" ")]
    data["4-gramme"][i] = list(ngrams(data["4-gramme"][i], 4))
data
```

```python
data2 = open("Natural_Language_Processing_Text.txt")
```

```python
data2 = data2.read()
```

```python
# convertir text en minuscul 
data2 = data2.lower()
data2 
```

```python
# Suppression des stop words
from nltk.corpus import stopwords
stopwords_en = set(stopwords.words('english'))
data2 = data2.split(" ")
data2 = [w for w in data2 if w not in stopwords_en]
data2 = " ".join(data2)
data2
```

```python
# tokenisation
data2 = data2.split(".")
for i in range(len(data2)):
    data2[i] = data2[i].strip()
data2
```

```python
def n_gramme(data, n):
    data_n_gramme = [w for w in data.split(" ")]
    data_n_gramme = list(ngrams(data_n_gramme,n))
    return data_n_gramme
```

```python
data2_2_gramme = []
for i in range(len(data2)):
    data2_2_gramme.append(n_gramme(data2[i],2))
data2_2_gramme
```

```python
data2_3_gramme = []
for i in range(len(data2)):
    data2_3_gramme.append(n_gramme(data2[i],3))
data2_3_gramme
```

```python
data2_4_gramme = []
for i in range(len(data2)):
    data2_4_gramme.append(n_gramme(data2[i],4))
data2_4_gramme
```

```python
# TF_IDF
from sklearn.feature_extraction.text import TfidfVectorizer
tf_idf_vect = TfidfVectorizer()
data2_tf_idf = []
tf_idf_vect.fit(data2)
data2_tf_idf.append(tf_idf_vect.transform(data2).toarray())
data2_tf_idf
```

```python
import pandas as pd
data3 = pd.read_csv("train_rel_2.tsv",delimiter="\t")
```

```python
data3
```

```python
# Convertir en lower
for i in range(len(data3)):
    data3["EssayText"][i] = data3["EssayText"][i].lower()
data3
```

```python
# Supprimer les ponctuations et lemmatization
import string
import nltk
from nltk.stem import WordNetLemmatizer
ponc = string.punctuation
wnl = WordNetLemmatizer()
for i in range(len(data3)):
    data3["EssayText"][i] = nltk.word_tokenize(data3["EssayText"][i])
    data3["EssayText"][i] = [w for w in data3["EssayText"][i] if w not in ponc]
    data3["EssayText"][i] = [wnl.lemmatize(w) for w in data3["EssayText"][i]]
    data3["EssayText"][i] = " ".join(data3["EssayText"][i])
data3
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
tf_idf_vect = TfidfVectorizer()
data3_tf_idf = []
tf_idf_vect.fit(data3["EssayText"])
data3_tf_idf.append(tf_idf_vect.transform(data3["EssayText"]).toarray())
data3_tf_idf
```

```python
data44 = pd.read_json('News_Category_Dataset.json', lines=True)
```

```python
data44
```

```python
# Supprimer les ponctuations, convertir en lower, supprimer les stop words & stemming
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
stm = PorterStemmer()
stpw = set(stopwords.words('english'))
ponc = string.punctuation
for i in range(1000):
    data44["headline"][i] = data44["headline"][i].lower()
    data44["short_description"][i] = data44["short_description"][i].lower()
    data44["headline"][i] = nltk.word_tokenize(data44["headline"][i])
    data44["short_description"][i] = nltk.word_tokenize(data44["short_description"][i])
    data44["headline"][i] = [w for w in data44["headline"][i] if w not in ponc]
    data44["short_description"][i] = [w for w in data44["short_description"][i] if w not in ponc]
    data44["headline"][i] = [w for w in data44["headline"][i] if w not in stpw]
    data44["short_description"][i] = [w for w in data44["short_description"][i] if w not in stpw]
    data44["headline"][i] = [stm.stem(w) for w in data44["headline"][i]]
    data44["short_description"][i] = [stm.stem(w) for w in data44["short_description"][i]]
    data44["headline"][i] = " ".join(data44["headline"][i])
    data44["short_description"][i] = " ".join(data44["short_description"][i])
data44
```

```python
# Bag of words
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
short_description = []
headline = []
for i in range(1000):
    short_description.append(data44["short_description"][i])
    headline.append(data44["headline"][i])
vectorizer.fit(short_description)
vectorizer.fit(headline)
short_description_bag_of_words = vectorizer.transform(short_description)
headline_bag_of_words = vectorizer.transform(headline)
```
