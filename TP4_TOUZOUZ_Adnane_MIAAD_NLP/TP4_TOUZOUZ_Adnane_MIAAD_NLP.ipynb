```python
import pandas as pd
data1 = pd.read_csv("Tweets.csv")
```

```python
# On vas travailler avec le feature text on vas faire le prÃ©traitement
# Supprimer les Ponctuations, tokinisation, convertir en miniscul, supprimer les stopWords & stemming
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
stem = PorterStemmer()
stopwords_en = set(stopwords.words('english'))
ponc = string.punctuation
for i in range(len(data1)):
    data1["text"][i] = data1["text"][i].lower()
    data1["text"][i] = nltk.word_tokenize(data1["text"][i])
    data1['text'][i] = [w for w in data1['text'][i] if w not in ponc]
    data1["text"][i] = [w for w in data1["text"][i] if w not in stopwords_en]
    data1["text"][i] = [stem.stem(w) for w in data1["text"][i]]
    data1["text"][i] = " ".join(data1["text"][i])
data1
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
tf_idf_vect = TfidfVectorizer()
data1_text = []
data1_airline_sentiment = []
tf_idf_vect.fit(data1["text"])
data1_text=tf_idf_vect.transform(data1["text"]).toarray()
tf_idf_vect.fit(data1["airline_sentiment"])
data1_airline_sentiment=tf_idf_vect.transform(data1["airline_sentiment"]).toarray()
data1_airline_sentiment
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
```
